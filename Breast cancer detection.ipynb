```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier

import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Cancer detection dataset.csv')
pd.options.display.max_columns = None
```

```python
new_df = df.copy()
```

```python
new_df.head()
```

```python
new_df = new_df.drop(['id'], axis=1)
```

```python
new_df = new_df.dropna()
```

```python
new_df.diagnosis.value_counts()
```

```python
diagnosis = df.diagnosis.astype('category').cat.codes
```

```python
new_df = pd.concat([new_df, diagnosis], axis=1)
```

```python
new_df.columns.values
```

```python
columns = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean',
       'concavity_mean', 'concave_points_mean', 'symmetry_mean',
       'fractal_dimension_mean', 'radius_se', 'texture_se',
       'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se',
       'concavity_se', 'concave_points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave_points_worst',
       'symmetry_worst', 'fractal_dimension_worst', 'Diagnosis']
```

```python
new_df.columns = columns
```

```python
new_df = new_df.drop(['diagnosis'], axis=1)
```

```python
new_df.head()
```

```python
clean_df = new_df.copy()
```

```python
target = clean_df.Diagnosis
featuers = clean_df.iloc[:,:-1]
```

```python
scaler = StandardScaler()
std_df = scaler.fit_transform(featuers)
```

```python
x_train,x_test,y_train,y_test = train_test_split(featuers, target, random_state = 42)
```

```python
x_train_np = np.array(x_train)
x_test_np = np.array(x_test)
y_train_np = np.array(y_train)
y_test_np = np.array(y_test)
```

```python
clf = MLPClassifier(solver = 'lbfgs', alpha = 1e-5,
                   hidden_layer_sizes = (50,50), random_state = 42)
```

```python
clf.fit(x_train_np,y_train_np)
```

```python
y_pred = clf.predict(x_test_np)
```

```python
confusion_matrix(y_test_np, y_pred)
```

```python
y_test.value_counts()
```

```python
y_pred = np.reshape(y_pred, (-1,1))
y_test_np = np.reshape(y_test_np,(-1,1))
```

```python
cross_score = cross_val_score(clf,y_test_np,y_pred, cv=5)
```

```python
print(cross_score)
print('mean: ',np.mean(cross_score))
```

```python
pickle.dump(clf, open ('clf.pickle', 'wb'))
```
